## example for use quicksort

To identify the top-performing stocks based on percentage return, you can apply the following mathematical steps using Quick Sort:

Create a List: First, create a list of these companies, where each element of the list is a tuple containing the company's name and its percentage return.

before sorted
data = [    ('Company A', 12),    ('Company B', 8),    ('Company C', 15),    ('Company D', 10),    ('Company E', 9)]

after sorted
sorted_data = [    ('Company C', 15),    ('Company A', 12),    ('Company D', 10),    ('Company E', 9),    ('Company B', 8)]



```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[len(arr) // 2]
        left = [x for x in arr if x < pivot]
        middle = [x for x in arr if x == pivot]
        right = [x for x in arr if x > pivot]

        return quick_sort(left) + middle + quick_sort(right)

# Example usage:
my_list = [3, 6, 8, 10, 1, 2, 1]
sorted_list = quick_sort(my_list)
print(sorted_list)  # Output should be [1, 1, 2, 3, 6, 8, 10]
```

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)

# Take user input to create a list
user_input = input("Enter a list of integers separated by spaces: ")
my_list = [int(x) for x in user_input.split()]

# Sort the list using quick_sort
sorted_list = quick_sort(my_list)
print("Sorted list:", sorted_list)
```
